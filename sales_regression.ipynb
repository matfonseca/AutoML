```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(dataset_venta, test_size=0.3, random_state=1)
```

```python
train.info()
```

```python
from pycaret.regression import * 
```

```python
train.columns
```

```python
numerical_features = list(train.columns)
numerical_features.remove('dolar_price')
numerical_features.remove('ambientes')

categorical_feature = ["ambientes"]

```

```python
exp_1001 = setup(data=train, target='dolar_price', 
                 numeric_features=numerical_features,
                 categorical_features=categorical_feature,
                 normalize=True,fold = 10,
                 log_experiment=True, experiment_name='dataset_ventas_1002',
                 session_id = 1002)
```

```python
best_model = compare_models()
```

## Xgboost

```python
xgboost = create_model('xgboost')
```

```python
tuned_xgboost_model = tune_model(xgboost, optimize = 'R2')
```

```python
plot_model(tuned_xgboost_model, plot = 'feature')
```

```python
plot_model(tuned_xgboost_model, plot = 'error')
```

```python
interpret_model(tuned_xgboost_model)
```

```python
predict_model(tuned_xgboost_model)
```

```python
final_xgboost_model = finalize_model(tuned_xgboost_model)
```

```python
save_model(final_xgboost_model,'./models/202011110_final_xgboost_model')
```
